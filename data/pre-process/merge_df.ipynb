```python
import pandas as pd
```

```python
pop_df = pd.read_csv('population/Data of Population 2002 to 2023.csv')
pop_df = pop_df[[ 'ADM1_EN','Month','Year', 'Population']]
pop_df =pop_df.rename({'ADM1_EN':'province','Month':'month','Year':'year','Population':'population'},axis=1)
pop_df = pop_df.replace({'Loburi': 'Lop Buri','Buang Kan':'Nong Khai'})
pop_df = pop_df.groupby(['province','month','year']).sum().reset_index()
pop_df.head(5)
```

```python
climate_df = pd.read_csv('terraclimate/climate_data_terra_2003_2023_bio_76prov.csv')
climate_df = climate_df[['province','month','year',]+[x for x in climate_df.columns if x not in ['province','month','year']]]
climate_df = climate_df.replace({'Loburi': 'Lop Buri','Buang Kan':'Nong Khai'})
climate_df = climate_df.groupby(['province','month','year']).mean().reset_index()
climate_df.head(5)
```

```python
dengue_df = pd.read_csv('dengue/combine_dengue_province_monthly_77provinces.csv')
dengue_df = dengue_df.rename({'province':'province','count':'dengue_cases'},axis=1)
dengue_df = dengue_df[['province','month','year','dengue_cases',]]
dengue_df = dengue_df.replace({'Loburi': 'Lop Buri','Bueng Kan':'Nong Khai'})
dengue_df = dengue_df[dengue_df['year']!=2024]
dengue_df.head(5)
```

```python
merged_df = pd.merge(dengue_df,climate_df,on=['province','month','year'],how='left')
merged_df = pd.merge(merged_df,pop_df,on=['province','month','year'],how='left')
merged_df['dengue_rate'] = (merged_df['dengue_cases'] / merged_df['population']) * 100000
merged_df = merged_df.sort_values(['province','year','month'])
merged_df = merged_df[['province','month','year','dengue_cases', 'dengue_rate', ] + [x for x in merged_df.columns if x not in ['province','month','year','dengue_cases', 'dengue_rate',]]]
merged_df.to_csv('../dataset/historical_combine_dengue_CaseRate_terra_monthly_76provinces_population.csv',index=False)
```
